```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnx = sqlite3.connect('database.sqlite')
df = pd.read_sql_query("SELECT * FROM Player_Attributes", cnx)
print("done")
```

```python
filtered_df = df[['player_fifa_api_id','preferred_foot', 'dribbling', 'agility', 'balance', 'vision', 'positioning', 'ball_control','gk_reflexes' ]].copy()
```

```python
gk_filter = filtered_df['gk_reflexes'] < 20 
filtered_df = filtered_df[gk_filter]
filtered_df
```

```python
filtered_df.isnull().any()
```

```python
left_foot_filter = filtered_df['preferred_foot']=='left' 
```

```python
filtered_df_left = filtered_df[left_foot_filter]
```

```python
filtered_df_left
```

```python
right_foot_filter = filtered_df['preferred_foot']=='right' 
```

```python
filtered_df_right = filtered_df[right_foot_filter]
```

```python
filtered_df_right
```

```python
filtered_df_left.groupby('player_fifa_api_id')
```

```python
filtered_df_right.groupby('player_fifa_api_id')
```

```python
filtered_df_left
```

```python
filtered_df_right
```

```python
filtered_df_left_mean_by_player = filtered_df_left[['player_fifa_api_id', 'dribbling','agility', 'balance', 'vision', 
                                                    'ball_control']].groupby('player_fifa_api_id', as_index=False).mean()
```

```python
filtered_df_left_mean_by_player
```

```python
filtered_df_right_mean_by_player = filtered_df_right[['player_fifa_api_id', 'dribbling','agility', 'balance', 'vision', 
                                                    'ball_control']].groupby('player_fifa_api_id', as_index=False).mean()
```

```python
filtered_df_right_mean_by_player
```

```python
del filtered_df_left_mean_by_player['vision']
```

```python
del filtered_df_right_mean_by_player['vision']
```

```python
filtered_df_left_mean_by_player
```

```python
final_left = filtered_df_left_mean_by_player.drop(columns='player_fifa_api_id').mean(axis=1)
```

```python
final_left.describe()
```

```python
final_right = filtered_df_right_mean_by_player.drop(columns='player_fifa_api_id').mean(axis=1)
```

```python
final_right.describe()
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

```python
plt.hist(final_left, 50, density=False, facecolor='#87CEEB')

plt.xlabel('Ability')
plt.ylabel('Number of Players')
plt.title('Left Foot Ability')
plt.xlim([30, 100])
fig = plt.gcf()
fig.set_size_inches(15, 10)
fig.savefig('left.png', dpi=500)
#plt.grid(True)

plt.show()
```

```python
plt.hist(final_right, 50, density=False, facecolor='#87CEEB')

plt.xlabel('Ability')
plt.ylabel('Number of Players')
plt.title('Right Foot Ability')
plt.xlim([30, 100])
fig = plt.gcf()
fig.set_size_inches(15, 10)
fig.savefig('right.png', dpi=500)

plt.show()
```
